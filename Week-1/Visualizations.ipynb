```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('/content/student-mat.csv',sep=';')
df.head()
```

# pie chart

```python
# shown student's father's job , and most of them work as services
color = sns.color_palette('pastel')
plt.figure(figsize = (10,9))
plt.pie(df['Fjob'].groupby(df['Fjob']).count(),labels=df['Fjob'].unique() ,colors=color)
plt.show()
```

# Bar Chart

```python
# farmel = quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
# we understand from this plot that older student has good realationship whith there family
plt.figure(figsize = (10, 6))
plt.bar(df['famrel'],df['age'],color ='maroon',
        width = 0.3)
plt.xlabel("rating family relationships ")
plt.ylabel("student age")
```

```python
# reason : reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')
# most of the student chose there school according to their reputition and their coureses 
sns.barplot(data=df , x =df['reason'].groupby(df['reason']).count() , y=df['reason'].unique())
```

# box plot

```python
# G3 : final grade 
# most of student grade in range 7.5-14 from 20 
sns.boxplot(x=df['G3'])
```

```python
# absences : number of school absences (numeric: from 0 to 93)
# median  of student absences less than 10 times , but there are alot of outlair absences the whole class ! more than 70 times
plt.boxplot(df['absences'])
```

# scatter plot

```python
# health : current health status (numeric: from 1 - very bad to 5 - very good)
# acording to that we can see older student has very bad health 
sns.scatterplot(data=df, x=df['health'],y=df['age'])
```

```python
# studytime : weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
# older student dosent have much time for studing 
plt.scatter(x=df['studytime'],y=df['age'])
```

# Line Chart 

```python
#Dalc : workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
# older student drink alcohal more than younger 
sns.lineplot(data=df, x="Dalc", y="age")
```

```python
# Medu : mother's education(numeric: 0 - none, 1 - primary education (4th grade), 2 Ã¢â‚¬â€œ 5th to 9th grade, 3 Ã¢â‚¬â€œ secondary education or 4 Ã¢â‚¬â€œ higher
# higher mother education rearing there childern to take care about there health 
plt.plot(df['Medu'], df['health'])
plt.xlabel('mother education')
plt.ylabel('student health care')
plt.show()
```

# Histogram

```python
# males have so much freetime 
sns.histplot(data=df,x='freetime', hue="sex")
```

```python
# most of student are under age 19 
plt.hist(df['age'],color ='green')
plt.xlabel('AGE')
plt.ylabel('Count')
```

# Sub plot

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('favorite plots made')
ax1.bar(df['famrel'],df['age'],color ='maroon',
        width = 0.3)
ax2.scatter(x=df['studytime'],y=df['age'])
ax3.plot(df['Medu'], df['health'])
ax4.hist(df['age'],color ='green')
```
